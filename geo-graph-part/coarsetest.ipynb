## Testing uncoarsen

### Coarsest graph

```python
initialPartition = [random.randint(0,1) for i in range(len(g3.vs))]
            
cutEdges = []
for e in g3.es:
    if initialPartition[e.source] != initialPartition[e.target]:
        cutEdges.append(True)
    else:
        cutEdges.append(False)
            
color_dict_vertex = {0: "blue", 1: "red", 2: "green", 3: "pink", 4: "orange" }
visual_style = {}
layout = g3.layout("fr")
visual_style["layout"] = layout
visual_style["vertex_label"] = [v.index for v in g3.vs]
visual_style["vertex_size"] = 20
visual_style["bbox"] = (400, 400)
visual_style["margin"] = 100
visual_style["vertex_size"] = 20
visual_style["vertex_color"] = [color_dict_vertex[initialPartition[i]] for i in range(len(g3.vs))]
visual_style["edge_color"] = ["black" if cutEdges[i] else color_dict_vertex[initialPartition[g3.es[i].source]] for i in range(len(g3.es))]
ig.plot(g3, **visual_style)
```

### Uncoarsen once

```python
finerPartition = coarse.uncoarsen(g3, g2, initialPartition)
print(initialPartition)
print(finerPartition)

cutEdges = []
for e in g2.es:
    if finerPartition[e.source] != finerPartition[e.target]:
        cutEdges.append(True)
    else:
        cutEdges.append(False)
            
color_dict_vertex = {0: "blue", 1: "red", 2: "green", 3: "pink", 4: "orange" }
visual_style = {}
layout = g2.layout("fr")
visual_style["layout"] = layout
visual_style["vertex_label"] = [v.index for v in g2.vs]
visual_style["vertex_size"] = 20
visual_style["bbox"] = (400, 400)
visual_style["margin"] = 100
visual_style["vertex_size"] = 20
visual_style["vertex_color"] = [color_dict_vertex[finerPartition[i]] for i in range(len(g2.vs))]
visual_style["edge_color"] = ["black" if cutEdges[i] else color_dict_vertex[finerPartition[g2.es[i].source]] for i in range(len(g2.es))]
ig.plot(g2, **visual_style)
```
